```python
#생존자와 사망자의 나이는 어떻게 차이가 나는지 --> 이상값 제거
q3=df['age'].quantile(q=0.75)
q1=df['age'].quantile(q=0.25)

iqr=q3-q1

condition=df['age']<=4*iqr
new_df=df[condition]

sns.boxplot(data=new_df,x='survived',y='age')
#3분위 값을 q3,1분위 값을 q1로 설정해놓고
#iqr=q3-q1이라는 값으로 설정
#또한, iqr*4를 통해서 이상값 여부를 판단
```

```python
#생존자 수 vs 사망자 수
df['survived'].value_counts()
#사망자는 424명, 생존자는 290명으로 사망자가 더 많다는 것을 확인
```

```python
#각 컬럼별로, 어떤 상관관계가 존재할까?
sns.heatmap(df.corr(),annot=True)
#seaborn의 heatmap을 사용, annot=True를 통해서 각 히트맵마다 상관관계 사용
#색이 옅어질수록 강한 상관관계가 있다는 것을 확인
#5번질문(객실등급별로의 사망자수)과 엮어보면, 객실등급과 생존율은 -0.36으로, 약한 음의 상관관게가 있다고 할 수 있음.
#생존율은 반대로 1이 생존, 0이 사망
#객실등금이 올라갈수록(객실이 안좋아질수록) 생존률이 낮아지는 것으로 해석할 수 있음
```

```python
#생존률에 어느변수가 가장 큰 영향을 미쳤을까?
df.corr()['survived'].sort_values(ascending=False)
#가장 강한 상관관계는 -0.35로 약한 음의 상관관계를 갖고있는 pclass, 즉 객실등급이 가장 큰 영향을 미쳤다고 할 수 있음.
#fare --> 객실요금은 객실등급과도 연관이 되있으므로, 이 또한 상관관계가 있다고 볼 수 있음.
```
